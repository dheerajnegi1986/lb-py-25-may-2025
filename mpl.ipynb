# Matplotlib - visualization in python

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x, y)
plt.show()
```

```python
plt.title('Simple line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.plot(x, y)
plt.show()
```

```python
# OBJECT ORIENTED PLOT
# fig: entire figure, ax: individual plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Object oriented')
plt.show()
```

```python
x = [1,2,3,4]

y1 = [10,20,25,30]
y2 = [30,25,5,10]

fig, ax = plt.subplots(1,2) # 1 row, 2 col

fig.suptitle('My two plots')

ax[0].plot(x, y1, color='blue')
ax[1].plot(x, y2, color='red')

ax[0].set_title('First plot')
ax[1].set_title('Second plot')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(3,2), dpi=400)
plt.plot(x,y)
plt.title("Custom size & DPI")
plt.show()
```

```python
# save image locally
plt.plot(x,y)
plt.title('Save this plot to local')
plt.savefig('first-plot.png', dpi=300)
plt.close()
```

```python
# legend
plt.plot(x,y, label='Growth')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time v/s Value')
plt.legend()
plt.show()
```

```python
plt.plot(x,y, color='green', linewidth=2, linestyle='--')
plt.show()
```

```python
plt.plot(x,y, color='purple', linewidth=2, linestyle='--', marker='o', markersize=10)
plt.show()
```

```python
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,50)
plt.show()
```

```python
# linestyle -   --   -.   :  
# marker o s ^ 
plt.plot(x,y, color='purple', linewidth=2, linestyle='', marker='^', markersize=10)
plt.show()
```

```python
# bar graphs
plt.bar(['A', 'B', 'C'], [5,7,3], color=['skyblue', 'lightgreen', 'salmon'], hatch=['/', '\\', 'o'])
plt.title('Bar chart')
plt.show()
```

```python
# Scatter plot
plt.scatter(x,y, color='red')
plt.title('Scatter plot')
plt.show()
```

```python
# Histogram
data = [1,2,2,3,3,3,4,4,4,4,5,5]
plt.hist(data, bins=5)
plt.title('Histogram')
plt.show()
```

```python
# Pie chart
sizes = [30,45,25]
labels= ['A', 'B', 'C']
colors=['#ff9999','#66b3ff','#99ff99']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.show()
```
