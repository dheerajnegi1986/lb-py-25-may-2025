# Seaborn

Python data visualization library based on matplotlib.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
'''
Other dataset present
Tips
Iris
Penguins
Flights
Diamonds
.............
'''
flights = sns.load_dataset('flights')
flights.head()
```

```python
# Histplot
sns.histplot(data= tips, x='total_bill', bins=20, kde=True)
plt.title('Histogram of total bill with KDE')
plt.show()
```

```python
# KDEPLOT
sns.kdeplot(data=tips['tip'], fill=True, bw_adjust=0.5)
plt.title('KDE Plot of tip amount')
plt.show()
```

```python
# New dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Displot
sns.displot(
    data=iris,
    x='sepal_length',
    hue='species',
    kind='kde',
    fill=True
)

plt.title('Sepal length distribution by species')
plt.show()
```

```python
# Barplot
sns.barplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
)
plt.title('Average total bill by day and gender')
plt.show()
```

```python
# Boxplot
sns.boxplot(
    data=tips,
    x='day',
    y='tip',
    hue='sex'
)
plt.title('Boxplot of tip by day and gender')
plt.show()
```

```python
# Violinplot
sns.violinplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    split=True
)
plt.title('Violin plot of total bill')
plt.show()
```

```python
# Stripplot
sns.stripplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    jitter=True,
    dodge=True
)
plt.title('Raw tip data by day and gender')
plt.show()
```

```python
# Swarmplot
sns.swarmplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    dodge=True
)
plt.title('Swarmplot of bills')
plt.show()
```

```python
# Regplot
sns.regplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title('Total bill v/s tips regression line')
plt.show()
```

```python
# Lmplot
sns.lmplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    col='smoker'
)
plt.title('LM plot, Gender & smoker wise trend', y=1.1)
plt.show()
```

```python
# pairplot
sns.pairplot(
    data=iris,
    hue='species',
)
plt.show()
```
